```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt


tasks = {
    (0, 1): 3,
    (0, 2): 4,
    (1, 3): 1,
    (3, 4): 3,
    (1, 4): 3,
    (2, 4): 3
}


G = nx.DiGraph()
for (start, end), duration in tasks.items():
    G.add_edge(start, end, weight=duration)


pos = nx.spring_layout(G)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='skyblue')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Project Task Graph")
plt.show()


length, path = nx.single_source_dijkstra(G, source=0, target=4, weight='weight')
print("Critical Path:", path)
print("Total Duration (Critical Path):", length)


```
